## Exercise 5: Data Visualization in Jupyter Notebook

```python
%matplotlib inline
import pandas as pd
data_url="https://data.wprdc.org/datastore/dump/76fda9d0-69be-4dd5-8108-0de7907fc5a4"
complaints = pd.read_csv(data_url, parse_dates=True, index_col="CREATED_ON")
complaints.resample("M").size().plot();
```

- Based on the data in the plot, I see that there is a slight positive trend in the data with multiple modes and peaks. There also appears to be an increase in the number of complaints as the years go on
