```python



health_df2= {'Name': ['Frank', 'Grace'],'Age': [22, 29],'Weight': [62, 72]}



health_df1={'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
               'Age': [30, 25, 40, 35, 28],
               'Weight': [65, 70, 80, 75, 68]}


health_df1=pd.DataFrame(health_df1) 

health_df2=pd.DataFrame(health_df2) 
combined_data=pd.merge(health_df1,health_df2, how="outer")


weight_average=combined_data['Weight']

print("The mean of the new data is",weight_average.mean(),"\nThe new weight average is also smaller than that of the original dataframe")

```
