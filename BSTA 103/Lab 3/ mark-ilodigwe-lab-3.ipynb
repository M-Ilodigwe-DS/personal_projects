```python
#This box plots x, y, and z all together incorporating labels 
fig = plt.figure(figsize=(7,4), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])


ax2.plot(x,y)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("Zoom")

ax2.set_ylim([30,50])
ax2.set_xlim([20.0,22.0])

ax.plot(x,z)
ax.set_xlabel("X")
ax.set_ylabel("Z")
ax.set_ylim([0,10000])
ax.set_xlim([0,100])
```

### The Data
For these exercises, we'll utilize the famous Titanic dataset. In the Machine Learning section of the course, we might revisit this dataset to predict passenger survival rates. However, at present, our focus will be solely on visualizing the data using seaborn:

```python

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#This is a value counts plot for the eaborn library for sex 
sns.countplot(x = "sex", data = titanic,color='blue')
```

```python
#This cell creates a joint grib with a histogram and a scatter plot
joint_plot= sns.JointGrid(data=titanic, x='fare', y='age')

# Plot scatterplot and histogram
joint_plot.plot(sns.scatterplot, sns.histplot)
```

```python
#This cell is used to generate a boxplot 
fig, ax = plt.subplots(figsize=(12, 8))
sns.boxplot(x= "class", y="age", data=titanic, palette="rainbow",hue="alive")
```

```python
#This cell is used to give information about the titanic dataset 
titanic.info
titanic.head()
titanic.value_counts()
```

```python
#This box utilizes seaborn to create a heatmatt fo the correlation between various variables 
titanic_corr=titanic.corr()
#titanicplt = titanic.pivot_table(values=titanic.corr, index=titanic.columns,columns=titanic.columns)
ax = plt.axes()
sns.heatmap(titanic_corr, cmap="coolwarm",ax=ax)
ax.set_title("titanic.corr()")
```
